## 2 dimensional gradient descent

In this notebook, we work with a random generated dataset for a linear regression with 50 datapoints, the slope is a=1 and the intercept is b=1. We define a function for the mean squared error (MSE) loss and calculte it. To get a feeling how the loss is depending on the parameters a and b we plot the 3 dimensional loss landscape. Then we calculate the gradients of the loss w.r.t a and b and update the starting values of the parameters a and b (a=3, b=3). We minimize the loss by using gradient descent and updating the parameters with the learning rate eta for 20 steps. Finally we plot the change of the parameters and the loss in a 3d plot. #### Simulated data 

Here we simulate 50 datapoints for x and y, where y=a*x+b with slope a=1 and intercept b=1

```python
#################
## Simulated data 
n = 50
np.random.seed(42)
x = np.random.normal(0,1,n)
y = x + 1 + np.random.normal(0,1,n) 
```

Lets plot the data in a scatter plot and use the skleran library to determine and print the opimal values for the slope a and the intercept b. We see that because of the random noise the optimal values are close to a=1 and b=1 but they have a bit deviation. 

```python
plt.scatter(x=x,y=y)
plt.title("simulated data")
plt.xlabel("x")
plt.ylabel("y")

model = LinearRegression()
res = model.fit(x.reshape((len(x),1)), y)
predictions = model.predict(x.reshape((len(x),1)))
plt.plot(x, predictions)

plt.show()
print("intercept = ",res.intercept_,"solpe = ", res.coef_[0],)
```

Here we define the mean squared error (MSE) loss as a function and caluvalte the minimal loss with the the optimal values form the cell above.

```python
def loss(a,b):
  return 1/len(x)*(np.sum((a*x + b - y)**2))
```

Now lets plot the loss landscape for our parameters a and b in a 3d plot. We see that the loss function is shaped like a bowl. The contour lines at the bottom indicate positions of equal loss values. On the right side there is a zoom for the region a=b=3, we see that if we decrease a and b from there, the loss also decreases.

```python

def loss(a,b):
  return 1/len(x)*(np.sum((a*x + b - y)**2))

fig = plt.figure(figsize=(14,6))
ax = fig.add_subplot(1, 2, 1, projection="3d")
a = b = np.arange(-1, 3.0, 0.05)
A, B = np.meshgrid(a, b)
zs = np.array([loss(a,b) for a,b in zip(np.ravel(A), np.ravel(B))])
loss = zs.reshape(A.shape)

ax.plot_surface(A, B, loss, cmap="viridis", rstride=3, cstride=3,edgecolor='none',alpha=0.9)
ax.contour(A, B, loss, 20, cmap="viridis", linestyles="solid", offset=0)
ax.view_init(40, 290)

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('loss',rotation="vertical")
ax.set_title('loss function')
plt.tick_params(axis='both', which='major', labelsize=7)


def loss(a,b):
  return 1/len(x)*(np.sum((a*x + b - y)**2))

ax = fig.add_subplot(1, 2, 2, projection="3d")
a = b = np.arange(2.9, 3.1, 0.01)
A, B = np.meshgrid(a, b)
zs = np.array([loss(a,b) for a,b in zip(np.ravel(A), np.ravel(B))])
loss = zs.reshape(A.shape)

ax.plot_surface(A, B, loss, cmap="viridis", rstride=3, cstride=3,edgecolor='none',alpha=0.9)
ax.contour(A, B, loss, 20, cmap="viridis", linestyles="solid", offset=0)
ax.view_init(40, 290)

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('loss',rotation="vertical")
ax.set_title('loss function')
plt.tick_params(axis='both', which='major', labelsize=7)
```

## The gradients 

From the book we know that the gradiens are:

$$
  grad_a = \frac{\partial}{\partial a}\frac{1}{n}\sum_{i=1}^n(a x_i + b - y_i)^2 = \frac{2}{n}\sum_{i=1}^n(a x_i + b - y_i) x_i 
$$


$$
  grad_b = \frac{\partial}{\partial b}\frac{1}{n}\sum_{i=1}^n(a x_i + b - y_i)^2 = \frac{2}{n}\sum_{i=1}^n(a x_i + b - y_i)  
$$

Now we want to use the gradients to find the optimal values for a and b to minimize the MSE loss.

Here we define the formula for the gradients w.r.t a and b as a function to use them latter for the parameter update.




```python
def grad(a,b):
  y_hat = a*x + b
  n2 = 2.0 / len(x)
  return n2*np.sum((y_hat - y)*x) , n2*np.sum((y_hat - y))


def loss(a,b):
  return 1/len(y)*(np.sum((a*x + b - y)**2))
```

We start with the initial values a=3 and b=3 and calculate the gradients of the loss w.r.t a and b, then we use the update rule to get new (better) values for a and b. Here we repeat this for 20 steps with the learning rate eta = 0.1 and we already get very close to the optimal values (see cell above).

```python
eta = 0.1
a_ = 3
b_ = 3
loss_ = loss(a_,b_)
a_history = a_
b_history = b_
loss_history= loss_
for i in range(20):
    grad_a, grad_b  = grad(a_,b_)
    a_ = a_ - eta * grad_a
    b_ = b_ - eta * grad_b
    loss_= loss(a_,b_)
    a_history =np.append(a_history,a_) 
    b_history =np.append(b_history,b_) 
    loss_history =np.append(loss_history,loss_) 
    print("a = ", a_, "b = ", b_, "gard_a = ", grad_a, "gard_b = ", grad_b, "loss = ", loss_)
```

In this plot we again look at the loss landscape for our parameters a and b in a 3d plot. We see the development of the parameters a and b. We started with a=3 and b=3, then we rolled down the loss landscape and after 20 update steps we are almost in the minimum. 

```python

def loss(a,b):
  return 1/len(x)*(np.sum((a*x + b - y)**2))


fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(1, 1, 1, projection="3d")
ax.plot3D(a_history, b_history, loss_history, 'black',lw=2)
a = b = np.arange(-1, 3.0, 0.05)
A, B = np.meshgrid(a, b)
zs = np.array([loss(a,b) for a,b in zip(np.ravel(A), np.ravel(B))])
loss = zs.reshape(A.shape)

ax.plot_surface(A, B, loss, cmap="viridis", rstride=3, cstride=3,edgecolor='none',alpha=0.8)
ax.contour(A, B, loss, 20, cmap="viridis", linestyles="solid", offset=0)
ax.view_init(50, 290)

ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('loss',rotation="vertical")
ax.set_title('loss function')
plt.tick_params(axis='both', which='major', labelsize=7)
ax.scatter3D(a_history, b_history,loss_history,"black")
ax.text(a_history[0]-0.4, b_history[0], loss_history[0], "start", color='black',zorder=100)
ax.text(a_history[-1]-0.4, b_history[-1], loss_history[-1]-1.5, "after 20 steps", color='black',zorder=100)
```
