#### Creation of the softmax

```python
plt.style.use('default') # macht den hintergrund weg 
plt.rcParams.update({'font.size': 15}) #Macht grosse axen
xvals = np.linspace(-10,10,100)
plt.plot(xvals, tf.math.softplus(xvals).numpy())
plt.plot(xvals, tf.math.exp(xvals).numpy(), ls='-.')
plt.legend(['softplus','exp'])
plt.ylim(0,10)
plt.savefig('ch05_sp.pdf')
```
