```
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf

%matplotlib inline
plt.style.use('default')

print("TF  Version",tf.__version__)
np.random.seed(42)
tf.random.set_seed(42)
```

```
model_vgg=tf.keras.applications.vgg16.VGG16(include_top=True, weights='imagenet', input_shape=(224,224,3), pooling=None, classes=1000)
```

```
model_vgg.summary()
```

```
from urllib.request import urlopen
from PIL import Image
```

```
img = (Image.open(urlopen("https://raw.githubusercontent.com/tensorchiefs/dl_book/master/imgs/cow.jpg")))
plt.imshow(img)
plt.show()
img = (Image.open(urlopen("https://raw.githubusercontent.com/tensorchiefs/dl_book/master/imgs/cow.jpg")))
new_width  = 224
new_height = 224
img = img.resize((new_width, new_height), Image.ANTIALIAS)
plt.imshow(img)
plt.show()
img=np.array(img)
```

```
img=tf.keras.applications.vgg16.preprocess_input(img)
img.shape
```

```
pred=model_vgg.predict(np.expand_dims(img,axis=0))
```

```
tf.keras.applications.vgg16.decode_predictions(pred)
```
